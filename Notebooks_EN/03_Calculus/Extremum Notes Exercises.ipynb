# Key Formulas for Extremum Problems

## 1. Identifying Critical Points and Extrema

## What is an Extremum?
Extrema are the highest or lowest points on a function within a certain interval. They can be:

Local Maximum: A point higher than its immediate neighbors.
Local Minimum: A point lower than its immediate neighbors.
Global Maximum or Minimum: The highest or lowest point across the entire domain.

To determine the maximum or minimum values of a function
f(x):
1. Compute the derivative: $f'(x)$.
2. Solve 𝑓′(𝑥)=0f′(x)=0 to locate potential critical points.
3. Apply the second derivative test:
   - $f''(x) > 0$ Local minimum.
   - $f''(x) < 0$ Local maximum.
   - $f''(x) = 0$, The test is inconclusive.

## 2. Quadratic Functions and Vertex Formula
For $f(x) = ax^2 + bx + c$, the $extremum$ (vertex) is at:
$$
x = -\frac{b}{2a}
$$
- If $a > 0$, Minimum at the vertex.
- If $a < 0$, Maximum at the vertex.

## 3. Optimization Under Constraints
Scenario: Rectangle with a fixed perimeter $P$:
- Variables: Let the rectangle’s sides be $x$ and $y$.
- Constraint: $2x + 2y = P$.
- Area formula: 𝐴=𝑥⋅𝑦A=x⋅y.
- Substitute $y = \frac{P}{2} - x$
$$
A(x) = x \left(\frac{P}{2} - x\right)
$$
Find critical points of $A(x)$ to maximize the area.

## 4. Analyzing Rational Functions
For $f(x) = \frac{g(x)}{h(x)}$:
1. Differentiate using the quotient rule:
$$
f'(x) = \frac{g'(x)h(x) - g(x)h'(x)}{[h(x)]^2}
$$
2. Set $f'(x) = 0$ to identify critical points.
3. Evaluate these points to determine whether they correspond to maxima or minima.

## Problem 1 – Maximizing Profit

### Given:
The profit function:
$$P(u) = -2u^2 + 50u - 300$$
where $𝑢$ is the number of units sold. Find the value of $𝑢$ that maximizes profit.

### Solution:
1. **First,I find the derivative: $P(u)$:
   $$P'(u) = -4u + 50$$

2. **Then, I set the derivative equal to zero to find critical points**:
   Set $P'(u) = 0$:
   $$-4u + 50 = 0$$
   $$u = \frac{50}{4} = 12.5$$

3. **Next, I used the second derivative test to confirm if it’s a maximum**:
p′′(u)=−4(Since P ′′(u)<0, it’s a maximum.)

### Answer:
The profit is maximized when $𝑢$=12.5 units are sold.

## Problem 4: Finding Extrema of $f(x) = \frac{x^2 + 2x + 1}{x - 1}$

### Given:
The rational function:
$$f(x) = \frac{x^2 + 2x + 1}{x - 1}$$

### Steps:
1. **Find the derivative using the quotient rule:**
   $$f'(x) = \frac{(2x + 2)(x - 1) - (x^2 + 2x + 1)(1)}{(x - 1)^2}$$

2. **Simplify the numerator**:
   $$f'(x) = \frac{2x^2 - 2x + 2x - 2 - (x^2 + 2x + 1)}{(x - 1)^2}$$
   $$f'(x) = \frac{x^2 - 2x - 3}{(x - 1)^2}$$

3. Set $f'(x) = 0$, which means solving the numerator:
   $$x^2 - 3 = 0$$
   Factorize:
   $$(x - 3)(x + 1) = 0$$
   $$x = 3, \, x = -1$$

### Answer:
Critical points are at $f(x) = \sqrt{3} \text{ and } $ $f(x) = -\sqrt{3}$ Further analysis is required to classify them as maxima or minima.
