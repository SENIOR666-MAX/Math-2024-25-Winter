# Functions




## Introduction to Functions

A function is a relation between a set of inputs and a set of outputs, where each input is related to exactly one output. Functions are fundamental in mathematics and programming, and they help to structure and simplify computations.

### Key Concepts in Mathematics:
- **Domain**: The set of all possible inputs.
- **Range**: The set of all possible outputs.
- **Function Notation**: \( f(x) \), where \( x \) is the input and \( f(x) \) is the output.

### Key Concepts in Programming:
- A function is a block of reusable code designed to perform a specific task.
- Functions are defined using the `def` keyword in Python.

This notebook provides examples of mathematical and programming functions.



## Mathematical Functions

### Linear Function
A linear function has the form:
\[
f(x) = mx + c
\]
Where:
- \( m \) is the slope.
- \( c \) is the y-intercept.

### Quadratic Function
A quadratic function has the form:
\[
f(x) = ax^2 + bx + c
\]

### Example
For \( f(x) = 2x + 3 \), calculate \( f(5) \):
\[
f(5) = 2(5) + 3 = 13
\]



## Programming Functions

In Python, a function is defined using the `def` keyword.

### Syntax
```python
def function_name(parameters):
    # Function body
    return result
```

### Example: Linear Function in Python
```python
def linear_function(x, m, c):
    return m * x + c

# Using the function
result = linear_function(5, 2, 3)  # f(5) = 2*5 + 3
print(result)  # Output: 13
```


````python

# Plotting a Quadratic Function
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def quadratic_function(x):
    return 2 * x**2 + 3 * x + 1

# Generate x values
x = np.linspace(-10, 10, 400)
y = quadratic_function(x)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="f(x) = 2x^2 + 3x + 1")
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.title("Quadratic Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
````

````python

# Example: Defining a Python Function
def greet(name):
    return f"Hello, {name}!"

# Using the function
print(greet("Alice"))  # Output: Hello, Alice!
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f(x):
    return x**2

def g(x):
    return np.sqrt(x)

def h(x):
    return 1 / x

def j(x):
    return np.sin(x)

# Define the x values
x = np.linspace(0.1, 5, 400)  # Avoid 0 for h(x) as it will cause division by zero

# Plot all functions
plt.figure(figsize=(10, 8))

# Plot f(x) = x^2
plt.plot(x, f(x), label=r"$f(x) = x^2$")

# Plot g(x) = x
plt.plot(x, g(x), label=r"$g(x) = \sqrt{x}$")

# Plot h(x) = 1/x
plt.plot(x, h(x), label=r"$h(x) = \frac{1}{x}$")

# Plot j(x) = sin(x)
x_sin = np.linspace(0, 5, 400)  # Wider range for sine
plt.plot(x_sin, j(x_sin), label=r"$j(x) = \sin(x)$")

# Add labels, title, and legend
plt.title("Graphs of f(x), g(x), h(x), and j(x)", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.legend(fontsize=12)
plt.grid()

# Show the plot
plt.show()

# Evaluate the functions at x = 2
x_val = 2
f_val = f(x_val)
g_val = g(x_val)
h_val = h(x_val)
j_val = j(x_val)

# Print the results
print(f"f(2) = {f_val} = 2^2 = 4")
print(f"g(2) = {g_val:.2f} = 2 = 1.41")
print(f"h(2) = {h_val:.2f} = 1/2 = 0.5")
print(f"j(2) = {j_val:.3f} = sin(2) = 0.909")

# https://www.geogebra.org/calculator/ynrz3wv7
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f(x):
    return 3 * x - 1

def g(x):
    return np.sqrt(x)

# Composite functions
def f_of_g(x):
    return f(g(x))  # f(g(x)) = 3x - 1

def g_of_f(x):
    return g(f(x))  # g(f(x)) = (3x - 1)

def f_of_f(x):
    return f(f(x))  # f(f(x)) = 3(3x - 1) - 1 = 9x - 4

def g_of_g(x):
    return g(g(x))  # g(g(x)) = x = x^(1/4)

# Define the x values for plotting
x = np.linspace(0.1, 10, 400)  # Start from 0.1 to avoid sqrt(negative)
x_valid_f = np.linspace(1/3, 10, 400)  # For g(f(x)), ensure 3x - 1 >= 0

# Plot the functions
plt.figure(figsize=(12, 8))

# Plot f(g(x))
plt.plot(x, f_of_g(x), label=r"$f(g(x)) = 3\sqrt{x} - 1$")

# Plot g(f(x))
plt.plot(x_valid_f, g_of_f(x_valid_f), label=r"$g(f(x)) = \sqrt{3x - 1}$")

# Plot f(f(x))
plt.plot(x, f_of_f(x), label=r"$f(f(x)) = 9x - 4$")

# Plot g(g(x))
plt.plot(x, g_of_g(x), label=r"$g(g(x)) = x^{1/4}$")

# Add labels, legend, and title
plt.title("Visualization of Composite Functions", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.legend(fontsize=12)
plt.grid()

# Show the plot
plt.show()

# Evaluate the composite functions at a specific point, e.g., x = 4
x_val = 4
print(f"a) f(g(x)) at x = {x_val} is {f_of_g(x_val):.2f}")
print(f"b) g(f(x)) at x = {x_val} is {g_of_f(x_val):.2f}")
print(f"c) f(f(x)) at x = {x_val} is {f_of_f(x_val):.2f}")
print(f"d) g(g(x)) at x = {x_val} is {g_of_g(x_val):.2f}")

# https://www.geogebra.org/calculator/buan48ub
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f(x):
    return np.exp(x)  # f(x) = e^x

def g(x):
    return np.log(x)  # g(x) = ln(x)

# Composite functions
def f_of_g(x):
    return f(g(x))  # f(g(x)) = e^(ln(x)) = x

def g_of_f(x):
    return g(f(x))  # g(f(x)) = ln(e^x) = x

# Define the x values for plotting
x_positive = np.linspace(0.1, 10, 400)  # x > 0 for g(x) and f(g(x))
x_all = np.linspace(-2, 2, 400)  # Includes negative values for f(x) and g(f(x))

# Plot the functions and their compositions
plt.figure(figsize=(12, 8))

# Plot f(g(x)) = x
plt.plot(x_positive, f_of_g(x_positive), label=r"$f(g(x)) = e^{\ln(x)} = x$")

# Plot g(f(x)) = x
plt.plot(x_all, g_of_f(x_all), label=r"$g(f(x)) = \ln(e^x) = x$")

# Add labels, legend, and title
plt.title("Composite Functions: f(g(x)) and g(f(x))", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.legend(fontsize=12)
plt.grid()

# Show the plot
plt.show()

# Evaluate the composite functions at a specific point
x_val = 2
f_g_val = f_of_g(x_val)  # f(g(x))
g_f_val = g_of_f(x_val)  # g(f(x))

# Print the results
print(f"f(g({x_val})) = {f_g_val}")
print(f"g(f({x_val})) = {g_f_val}")
````
